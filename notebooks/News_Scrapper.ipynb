```python
import pandas as pd
import numpy as np
import sys
import requests
from newspaper import Article
```

```python
urls = ['https://fxn.ws/2R55o6y', 'https://fxn.ws/2R0ZbrU', 'https://fxn.ws/343iIh0', 'https://fxn.ws/2w5N2eg', 
        'https://fxn.ws/39CjmDm', 'https://fxn.ws/2UDHvDS', 'https://fxn.ws/3dyoaMY', 'https://fxn.ws/33OZdZk',
        'https://fxn.ws/3abiVRt', 'https://fxn.ws/3dcpiG5']
```

```python
def get_news(url):
    
    try:
        #print(url)
        news = Article(url, language="en")
        news.download()
        news.parse() 
        news.nlp()
        d = {'author': news.authors, 'date': news.publish_date, 'title': news.title, "text": news.text, 
             "summary": news.summary, "keyword": news.keywords, 'url': news.url}
        return d
        
    except Exception as e:
        return None

def get_news_articles(urls):
    
    news_articles = dict()
    for i, url in enumerate(urls):
        news = get_news(url)
        if(news != None):
            news_articles[i] = news
            
    return news_articles
```

```python
news_articles = get_news_articles(urls)
```

```python
df = pd.DataFrame(columns = news_articles[0].keys())
for i in news_articles:
    df = df.append(news_articles[i], ignore_index = True)
df.shape       
```
